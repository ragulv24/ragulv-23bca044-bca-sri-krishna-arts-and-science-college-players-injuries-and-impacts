```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram using Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=15, color='skyblue', edgecolor='black', alpha=0.7)  # Changed 'data' to 'df'
plt.title('Histogram of Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Histogram using Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(
    data=df,  # Changed 'data' to 'df'
    x='fifa_rating',  # Replace with your desired column
    bins=15,
    kde=True,  # Add a Kernel Density Estimate curve
    color='coral',
    alpha=0.8
)
plt.title('Histogram of FIFA Rating with KDE', fontsize=16)
plt.xlabel('FIFA Rating', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# KDE plot using Seaborn
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=df,  # Changed 'data' to 'df' to use the DataFrame you created
    x='fifa_rating',  # Replace with your desired column
    shade=True,  # Adds shading under the curve
    color='blue',
    alpha=0.8,
    bw_adjust=0.5  # Bandwidth adjustment for smoothing
)
plt.title('KDE Plot of FIFA Rating', fontsize=16)
plt.xlabel('FIFA Rating', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Optional: KDE for multiple groups (e.g., 'position')
plt.figure(figsize=(10, 6))
sns.kdeplot(
    data=df,  # Changed 'data' to 'df' here as well
    x='age',  # Replace with your desired column
    hue='position',  # Group by a categorical column
    fill=True,  # Adds shaded areas for each group
    palette='viridis',
    alpha=0.6,
    common_norm=False  # Prevents normalizing across groups
)
plt.title('KDE Plot of Age by Position', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend(title='Position', fontsize=10, loc='upper right')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
